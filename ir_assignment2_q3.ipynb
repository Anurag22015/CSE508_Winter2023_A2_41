```python
#Maximum DCG score is calculate for the sorted dataset.
import pandas as pd
import itertools
import math

def calculateDCG(df,l):
  sum=df.iloc[0][0]
  for i in range(1,l):
      sum+=float(float(df.iloc[i][0])/math.log2(i+1))
  return sum
  

maxdcg=calculateDCG(df,len(df))
print("Maximum DCG score=", maxdcg)
print("Permutation corresponding to Maximum DCG")
print(sorted_df)
```

```python
count=[0]*4
for i in range(0,103):
  if df.iloc[i][0]==0:
    count[0]+=1
  if df.iloc[i][0]==1:
    count[1]+=1
  if df.iloc[i][0]==2:
    count[2]+=1
  if df.iloc[i][0]==3:
    count[3]+=1
print(count)
```

```python
#no of files number of such files that could be made should also be stated.
import math
n=(math.factorial(103)/(math.factorial(count[0])*math.factorial(count[1])*math.factorial(count[2])*math.factorial(count[3])))
n="{:.2f}".format(n)
print("NUmber of arrangement for the DCG are = ", n)
```

```python
#Normalized DCG at 50th position
nDCG=calculateDCG(df3,50)/calculateDCG(df,50)
print("Normalized DCG score=",nDCG)
```

```python
#Normalized DCG for whole datset
nDCG=calculateDCG(df3,len(df3))/maxdcg
print("Normalized DCG score=",nDCG)
```

```python
#calculate recall and precision for the doc
for i in range(0,103):
  df3.loc[i][75]=df3.iloc[i][75].split(":")[1]
  
```

```python
for i in range(0,103):
  print(df3.iloc[i][75])
```

```python
sorted_df3 = df3.sort_values(by=[75], ascending=False)
sorted_df3.head(20)
```

```python
for i in range(0,103):
  print(sorted_df3.iloc[i][75])
```

```python
total_relevant=0
for i in range(0,103):
  if float(sorted_df3.loc[i][75])==0:
    continue
  else:
    total_relevant+=1
print(total_relevant)
```

```python
retrieved_doc=1
relevant_retrived=0
precision=[]
recall=[]
for i in range(0,103):
  if int(sorted_df3.iloc[i][0])>0:
    relevant_retrived+=1
  recall.append(relevant_retrived/total_relevant)
  precision.append(relevant_retrived/retrieved_doc)
  retrieved_doc+=1
```

```python
print(recall)
```

```python
print(precision)
```

```python
import matplotlib.pyplot as plt
plt.plot(recall,precision)
plt.xlabel('recall')
plt.ylabel('precision')
plt.show()
```
