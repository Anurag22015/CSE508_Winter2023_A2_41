####****write pros and cons of each weighing scheme?****

####Binary (0,1) weighting:
#####Pros:
Simple and easy to understand.
Results in a compact and efficient representation of the document.
#####Cons:
Ignores the frequency of occurrence of terms in a document, which can be useful in determining the relevance of a document to a query.
Fails to distinguish between highly frequent and infrequent terms.
####Raw count (f(t,d)) weighting:
#####Pros:
Takes into account the frequency of occurrence of terms in a document, which can be useful in determining the relevance of a document to a query.
Easy to compute.
#####Cons:
Results in a sparse representation of the document, which can negatively affect the efficiency of search algorithms.
Fails to distinguish between highly frequent and infrequent terms.
####Term frequency (f(t,d)/Pf(t', d)) weighting:
#####Pros:
Normalizes the raw count by the maximum frequency of any term in the document to reduce the impact of highly frequent terms.
Considers the frequency of the term in the collection as well, which can help in distinguishing between rare and common terms.
#####Cons:
Assumes that all terms in the collection are equally important, which may not be true in all cases.
Can result in overly large weights for rare terms, which may lead to instability in the ranking.
####Log normalization (log(1+f(t,d))) weighting:
#####Pros:
Normalizes the raw count by taking the logarithm of the frequency, which reduces the impact of highly frequent terms.
Simple and easy to compute.
#####Cons:
Ignores the frequency of terms in the collection, which can result in rare terms receiving very low weights.
Can result in negative weights for very low-frequency terms, which may negatively affect the ranking.
####Double normalization (0.5+0.5*(f(t,d)/ max(f(t',d)))) weighting:
#####Pros:
Normalizes the raw count by taking into account the maximum frequency of any term in the document, which reduces the impact of highly frequent terms.
Normalizes the result of the previous step to reduce the impact of the document length, which can improve the ranking.
#####Cons:
Ignores the frequency of terms in the collection, which can result in rare terms receiving very low weights.
Requires more computation than some of the other weighting schemes.
